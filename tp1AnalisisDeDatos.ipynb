```python
import pandas as pd   #importamos la libreria pandas para poder leer el csv
import numpy as np
from ast import literal_eval
import re
import matplotlib.pyplot as plt

!file -i /content/dataset_adopciones.csv
df_adopciones = pd.read_csv('/content/dataset_adopciones.csv', encoding='latin-1')

df_adopciones.head()


df_adopciones['FECHA_INSCRIPCION'] = pd.to_datetime(df_adopciones['FECHA_INSCRIPCION'], format='%d%b%Y:%H:%M:%S')


df_adopciones['Año'] = df_adopciones['FECHA_INSCRIPCION'].dt.year


adopciones_por_año = df_adopciones['Año'].value_counts().sort_index()
print(adopciones_por_año)

plt.figure(figsize=(10, 6))
plt.plot(adopciones_por_año.index, adopciones_por_año.values, marker='o', linestyle='-', color='b')
plt.title('Adopciones por Año')
plt.xlabel('Año')
plt.ylabel('Cantidad de Adopciones')
plt.grid(True)
plt.show()

```

en este grafico de lineas que va del 2015 al 2024 te da a conocer que en el 2015 hubo menos de 100 chicos adoptados y los siguientes años superan los 200 chicos adoptados pero en la pandemia bajo ese numero hasta los 150 chicos adoptados y en los años posteriores se incremento la cifra hasta casi llegar a los 300 chicos adoptados.
en este 2024 hay muy poca gente adoptada porque recien empezo el año.

la conclusion final es que la gente adopta de una forma plena ante que la simple, la gente elige ambos generos y no hay un favorito y como ultima conclusion es que del año 2016 hasta el 2021 sacando el 2020 por la pandemia es que las cantidades de adopciones son muy igualadas.
